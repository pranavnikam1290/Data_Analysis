In order to segregate the salary into various categories below code is created

```python
def Income_bracket(x):
  if x <= 40000:
    return "Low"
  elif x > 40000 and x <= 70000:
    return "Medium"
  else:
    return "High"
```

```python
df['Income_Bracket'] = df['Income'].apply(Income_bracket)
```

```python
df
```

In order to segregate the Age into various categories below code is created

```python
def Age_bracket(x):
  if x <13:
    return "Kid"
  elif x >= 13 and x <= 19:
    return "Teen"
  elif x > 19 and x < 30:
    return "20s"
  elif x >= 30 and x < 40:
    return "30s"
  else:
    return "40s"
```

```python
df['Age_bracket'] = df['Age'].apply(Age_bracket)
```
